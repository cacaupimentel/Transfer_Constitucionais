```python
# Função para ler os arquivos das urls
def ler_arq_url(lcodmuni, lu, lmes):
  # Garda na variável a resposta para o método GET do site
  resposta = requests.get(lu)
  arq_json = resposta.json()
  # Transfora em dicionário
  objeto = dict()
  objeto = dict(arq_json)

  # Varrendo a collection
  for key,value in objeto.items():
    if key == "data":
      dados = value

      for k in dados:
        tmp = ""
        list_campos = []
        list_campos.append(lcodmuni)
        list_campos.append(k.get("ano"))
        list_campos.append(lmes)
        list_campos.append(k.get("nomefonte"))
        list_campos.append(k.get("categoria"))
        list_campos.append(k.get("nomecategoria"))
        list_campos.append(k.get("origem'"))
        list_campos.append(k.get("nomeorigem"))
        list_campos.append(k.get("especie"))
        list_campos.append(k.get("nomeespecie"))
        list_campos.append(k.get("rubrica"))
        list_campos.append(k.get("nomerubrica"))
        list_campos.append(k.get("alinea"))
        list_campos.append(k.get("nomealinea"))
        list_campos.append(k.get("subalinea"))
        list_campos.append(k.get("nomesubalinea"))
        list_campos.append(k.get("prevista"))
        list_campos.append(k.get("arrecadada"))
        # Grava arquivo
        tmp = lcodmuni + ";" + str(k.get("ano")) + ";" + str(lmes) + ";" + str(k.get("nomefonte")) + ";" + str(k.get("categoria")) + ";"
        tmp = tmp + str(k.get("nomecategoria")) + ";" + str(k.get("origem'")) + ";" + str(k.get("nomeorigem")) + ";" + str(k.get("especie")) + ";"
        tmp = tmp + str(k.get("nomeespecie")) + ";" + str(k.get("rubrica")) + ";" + str(k.get("alinea")) + ";" + str(k.get("nomealinea")) + ";"
        tmp = tmp + str(k.get("subalinea")) + ";"  + str(k.get("nomesubalinea")) + ";" + str(k.get("prevista")) + ";"
        tmp = tmp + str(k.get("arrecadada"))
        arq_csv.write(tmp + '\n')

        dataset_list.append(list_campos)

  #print(dataset_list)
  return dataset_list
```
